```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hot_hand = pd.read_csv('NBA Hot Hand Analysis(Sheet1) (1).csv')
hot_hand
```

```python
hot_hand['Hot Hand Three-Point Percentage'] = hot_hand['Hot Hand Three-Point Percentage'].fillna('0%')
hot_hand['Team Record'] = hot_hand['Team Record'].replace(', ', '-', regex=True)
hot_hand['Hot Hand Team Record'] = hot_hand['Hot Hand Team Record'].replace(', ', '-', regex=True)
hot_hand
```

```python
hot_hand['Field Goal Percentage'] = hot_hand['Field Goal Percentage'].str.replace('%', '').astype(float)
hot_hand['Hot Hand Field Goal Percentage'] = hot_hand['Hot Hand Field Goal Percentage'].str.replace('%', '').astype(float)

hot_hand['Three-point Percentage'] = hot_hand['Three-point Percentage'].str.replace('%', '').astype(float)
hot_hand['Hot Hand Three-Point Percentage'] = hot_hand['Hot Hand Three-Point Percentage'].str.replace('%', '').astype(float)

hot_hand['Free Throw Percentage'] = hot_hand['Free Throw Percentage'].str.replace('%', '').astype(float)
hot_hand['Hot Hand Free Throw Percentage'] = hot_hand['Hot Hand Free Throw Percentage'].str.replace('%', '').astype(float)

hot_hand
```

```python
raw_player_data = pd.read_csv('Hot Hand Game Data(Sheet1).csv')
raw_player_data
```

```python
raw_player_data['Blocks'] = raw_player_data['Blocks'].fillna(0)
raw_player_data['Steals'] = raw_player_data['Steals'].fillna(0)
raw_player_data = raw_player_data.drop('Team +/-', axis=1)
raw_player_data['Three Point Shots'] = raw_player_data['Three Point Shots'].fillna('0;0')
```

```python
raw_player_data = raw_player_data.rename(columns = {'Unnamed: 0': 'Player Name'})
```

```python
raw_player_data['Two Point Shots'] = raw_player_data['Two Point Shots'].replace(';', '/', regex=True)
raw_player_data['Three Point Shots'] = raw_player_data['Three Point Shots'].replace(';', '/', regex=True)
raw_player_data['Free Throws'] = raw_player_data['Free Throws'].replace(';', '/', regex=True)
```

```python
raw_player_data = raw_player_data.drop('Notes', axis=1)
```

```python
raw_player_data
```

I had to clean the 'raw_player_data' dataset because I had put some missing values in that I wanted replaced with 0s; I also decided to get rid of the 'Team +/-' column because it was unimportant.
```python
plt.figure(figsize=(8, 6))

plt.barh(['Points (per 100)', 'Hot Hand Points (per 100)'], 
        [pp100_mean, hhpp100_mean],
        color=['blue', 'orange'])

plt.title('Average Points Per 100 Possessions with and without Hot Hand', fontsize=16)
plt.ylabel('Normal vs. Hot Hand', fontsize=14)
plt.xlabel('Average Points Per 100 Possessions', fontsize=14)

plt.tight_layout()
plt.show()
```

The 20 players combined scored 75.4 more points per 100 possessions when they had the hot hand; that averages out to a 3.77 points per 100 possessions increase per player.

```python
x_labels = hot_hand['Players and Stats for 2023-24 Season']
x_pos = np.arange(len(hot_hand))

hot_hand.index = [f'Row {i+1}' for i in range(20)]

x = np.arange(len(hot_hand)) 
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))
bars1 = ax.bar(x - width/2, hot_hand['Field Goal Percentage'], width, label='FG% Without Hot Hand')
bars2 = ax.bar(x + width/2, hot_hand['Hot Hand Field Goal Percentage'], width, label='FG% With Hot Hand')


ax.set_xlabel('Players')
ax.set_ylabel('Field Goal Percentage')
ax.set_title('Comparison of FG% with and without the hot hand')
ax.set_xticks(x_pos)
ax.set_xticklabels(x_labels, rotation=75)
ax.legend()

plt.tight_layout()
plt.show()
```

14 of the 20 players increased their FG%.

```python
plt.boxplot([hot_hand['Three-point Percentage'], hot_hand['Hot Hand Three-Point Percentage']], labels=['Without Hot Hand', 'With Hot Hand'])
plt.title('Average Three Point Percentages')
plt.ylabel('Three Point Percentage')
plt.show()
```

Without the hot hand, players' 3pt% sit in the 35-42% range; when players do get the hot hand, those numbers increase to a 36-52% range. There are two outliers in the hot hand data, however. One player went 0/1 and another went 1/10, but the average sits much higher. There is also a high-side outlier for the hot hand as Tyrese Haliburton shot 60% on 30 attempts!

```python
plt.figure(figsize=(8, 6))
plt.scatter([1] * len(hot_hand['Free Throw Percentage']), hot_hand['Free Throw Percentage'], color='blue', label='Without Hot Hand', s=100) 
plt.scatter([2] * len(hot_hand['Hot Hand Free Throw Percentage']), hot_hand['Hot Hand Free Throw Percentage'], color='orange', label='With Hot Hand', s=100) 

plt.title('Free Throw Percentage vs. Hot Hand Free Throw Percentage', fontsize=16)
plt.xlabel('', fontsize=14)
plt.ylabel('Percentage', fontsize=14)

plt.legend()

plt.show()
```

There are four players shooting under 80% from the FT line both with and without the hot hand. With the hot hand, though, there are six players sitting at or above 90% from the line, while only two players without the hot hand do that.  

```python
hot_hand[['Hot Hand Win Percentage', 'Losses']] = hot_hand['Hot Hand Team Record'].str.split('-', expand=True)
```

```python
hot_hand = hot_hand.drop('Losses', axis=1)
hot_hand = hot_hand.drop('Hot Hand Team Record', axis=1)
hot_hand
```

```python
hot_hand['Team Wins'] = hot_hand['Team Record'].str.split('-', n=1).str[0]
```

```python
hot_hand['Team Wins']
```

```python
hot_hand['Team Losses'] = hot_hand['Team Record'].str.split('-', n=1).str[1]
```

```python
hot_hand['Team Wins'] = pd.to_numeric(hot_hand['Team Wins'], errors='coerce')
hot_hand['Team Losses'] = pd.to_numeric(hot_hand['Team Losses'], errors='coerce')
```

```python
hot_hand['Total Games'] = hot_hand['Team Wins'] + hot_hand['Team Losses']
```

```python
hot_hand['Win Percentage'] = hot_hand['Team Wins'] / hot_hand['Total Games']
```

```python
hot_hand = hot_hand.drop('Team Wins', axis=1)
hot_hand = hot_hand.drop('Team Losses', axis=1)
hot_hand = hot_hand.drop('Total Games', axis=1)
hot_hand = hot_hand.drop('Team Record', axis=1)
```

```python
hot_hand['Hot Hand Win Percentage'] = pd.to_numeric(hot_hand['Hot Hand Win Percentage'], errors='coerce')
```

```python
hot_hand['Hot Hand Win Percentage'] = hot_hand['Hot Hand Win Percentage'] / 10
```

```python
plt.figure(figsize=(12, 6))
hot_hand.set_index('Players and Stats for 2023-24 Season')[['Win Percentage', 'Hot Hand Win Percentage']].plot.area(alpha=0.6)

plt.title('Win Percentage when Player Gets the Hot Hand')
plt.xlabel('Players')
plt.ylabel('Win Percentage')
plt.ylim(0, 1)
plt.xlim(0, 20)
plt.grid(True)
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
